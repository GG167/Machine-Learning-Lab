```python

from sklearn.feature_extraction.text import CountVectorizer

count_vect=CountVectorizer(max_features=1500)
X=count_vect.fit_transform(df_train.text)

X_train_count=X.toarray()
X_train_count
```

```python
from sklearn.naive_bayes import BernoulliNB

y=df_train['target']

model=BernoulliNB()
model.fit(X,y)
```

```python
y_predict=model.predict(X)
y_predict
```

```python

from sklearn.metrics import accuracy_score
accu=accuracy_score(y,y_predict)
accu
```

```python
# Use TfidVectorizer instead of Countvectorizer and MultinomialNB

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

tfid_vec=TfidfVectorizer()
X_train = tfid_vec.fit_transform(df_train['text'])

y=df_train['target']

model=MultinomialNB()

model.fit(X_train,y)
```

```python
y_pred=model.predict(X_train)
y_pred
```

```python
accuracy=accuracy_score(y,y_pred)
accuracy
```
